```python
import json

fd = open("record.json",'r')
r = fd.read()
fd.close()

records = json.loads(r)
```

```python
records
```

```python
sale = {}
```

```python
prod_id = (input("Enter the product id: "))
prod_qnty = int(input("Enter the product quantity: "))
print("*********************")
print("Product name: ",records[prod_id]['product_name'])
print("Price: ",records[prod_id]['price'])
print("---------------------")
print("Amount: ",records[prod_id]['price']*prod_qnty)
print("*********************")

records[prod_id]['quantity'] = records[prod_id]['quantity'] - prod_qnty
js = json.dumps(records)

fd = open("record.json",'w')
fd.write(js)
fd.close()

sale[(len(sale) + 1)] = {'prod' : prod_id, 'qn' : prod_qnty, 'amount' : records[prod_id]['price']*prod_qnty}
js = json.dumps(sale)

fd = open("sale.json",'w')
fd.write(js)
fd.close()
```
